```python
def get_correlation(df):
#     price_cols = ['open', 'high', 'low', 'close']
#     indicator_df = df.drop(columns=price_cols)
    df = df.drop("timestamp" ,1)
#     df = df.sort_values(by=['timestamp'], ascending=False, ignore_index=True)

    corr_matrix = df.corr()

    return corr_matrix
#     print("HIGH: ", corr_matrix['high'].sort_values(ascending=False))
#     print("LOW: ", corr_matrix['low'].sort_values(ascending=False))
#     print("CLOSE: ", corr_matrix['close'].sort_values(ascending=False))
    
    
```

```python
five_data = five_df
five_data = get_correlation(five_data).head(4)
five_data = five_data.drop("open", 1)
five_data = five_data.drop("high", 1)
five_data = five_data.drop("low", 1)
five_data = five_data.drop("close", 1)
five_data = five_data.drop("volume", 1)
five_data.head()
```

```python
five_data = five_data.dropna(axis=1,how='all') #removes all null values 
five_data.head()
# five_data[five_data.threeoutside_value > 0]
```

```python
five_data = five_data.loc[:, (five_data > 0).any(axis=0)]
five_data.head(4)

```

```python
five_data = five_data.loc[:, ~(five_data <= 0).any()]
```
