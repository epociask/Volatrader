```python
%matplotlib
import plotly.graph_objects as go
import pandas as pd

df = pd.read_csv('ethusdt15m')
# df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
# df.head()
fig = go.Figure(data=go.Ohlc(x=df['timestamp'],
                    open=df['open'],
                    high=df['high'],
                    low=df['low'],
                    close=df['close']))
fig.show()
```
