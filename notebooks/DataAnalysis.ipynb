```python
import pandas as pd
df = pd.read_csv("ethusdt_data_5m")
df = df.drop("high", 1)
df = df.drop("low", 1)
rf = df.iloc[:, :-160]
rf.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = rf.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
import matplotlib.pyplot as plt

plt.show(ax)
```
